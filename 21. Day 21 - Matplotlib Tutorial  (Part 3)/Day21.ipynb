```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

## Styles with Matplotlib Plots

Matplotlib offers various styles to enhance the aesthetics of your plots. You can choose a predefined style or customize your own. This section will demonstrate how to apply styles to Matplotlib plots.






```python
# Applying a predefined style
plt.style.use('seaborn-darkgrid')

# Your plot code here
plt.plot(x, y, label='Data')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Styled Line Plot')
plt.legend()
plt.show()
```

## Adding Grid

Grid lines provide a visual reference in plots. You can add grid lines to both the x-axis and y-axis for better readability.





```python
# Adding grid lines
plt.plot(x, y, label='Data')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot with Grid')
plt.grid(True)  # Add grid lines
plt.legend()
plt.show()
```

## Handling Axes

You can customize the range and scale of the axes to focus on specific regions of interest in your data.




```python
# Customizing axes
plt.plot(x, y, label='Data')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot with Customized Axes')
plt.xlim(0, 6)  # Set x-axis limits
plt.ylim(0, 12)  # Set y-axis limits
plt.legend()
plt.show()
```

## Handling X and Y Ticks

Ticks represent the values on the axes. You can customize their positions and labels for better clarity.





```python
# Customizing ticks
plt.plot(x, y, label='Data')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot with Customized Ticks')
plt.xticks([1, 2, 3, 4, 5], ['One', 'Two', 'Three', 'Four', 'Five'])  # Customize x-axis ticks
plt.yticks(np.arange(0, 12, 2))  # Customize y-axis ticks
plt.legend()
plt.show()
```

## Adding Labels

Labels provide additional context to your plots. You can add labels to the x-axis, y-axis, and provide a title to the entire plot.




```python
# Adding labels
plt.plot(x, y, label='Data')
plt.xlabel('Time (s)')  # X-axis label
plt.ylabel('Velocity (m/s)')  # Y-axis label
plt.title('Velocity over Time')  # Plot title
plt.legend()
plt.show()
```


## Adding Title

A title provides a quick summary of the plot's purpose. It should be concise but informative.



```python
# Adding a title
plt.plot(x, y, label='Data')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot with Title')  # Add a title
plt.legend()
plt.show()
```
