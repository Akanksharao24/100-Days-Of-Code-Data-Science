## Sets

### 1. Introduction to Sets:
- A set is an unordered collection of unique elements.
- Sets do not allow duplicate values, and they are defined by enclosing comma-separated values within curly braces `{}`.
  ```python
  my_set = {1, 2, 3, 3, 4, 5}
  # Result: {1, 2, 3, 4, 5} (duplicates are automatically removed)
  ```



````python
my_set = {1, 2, 3, 3, 4, 5}
my_set
````

### 2. Set Operations:
- Sets support various operations common in mathematical set theory:
  - **Union (`|`):** Combines elements from two sets, excluding duplicates.
  - **Intersection (`&`):** Returns elements common to both sets.
  - **Difference (`-`):** Returns elements present in the first set but not in the second.
  - **Symmetric Difference (`^`):** Returns elements that are in either of the sets but not in both.
  
  ```python
  set1 = {1, 2, 3}
  set2 = {3, 4, 5}
  union_set = set1 | set2
  # Result: {1, 2, 3, 4, 5}
  intersection_set = set1 & set2
  # Result: {3}
  difference_set = set1 - set2
  # Result: {1, 2}
  symmetric_difference_set = set1 ^ set2
  # Result: {1, 2, 4, 5}
  ```



````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1 | set2
print('Union Set:',union_set)

intersection_set = set1 & set2
print('Intersection Set:',intersection_set)

difference_set = set1 - set2
print('Difference Set:',difference_set)

symmetric_difference_set = set1 ^ set2
print('Symmetric Difference Set:',symmetric_difference_set)
````

### 3. Set Methods:
- Sets come with several built-in methods for manipulation:
  - `add(x)`: Adds element `x` to the set.
  - `remove(x)`: Removes element `x` from the set. Raises an error if `x` is not present.
  - `discard(x)`: Removes element `x` from the set if it is present. Does not raise an error if `x` is not found.
  - `pop()`: Removes and returns an arbitrary element from the set.
  - `clear()`: Removes all elements from the set.
  
  ```python
  my_set = {1, 2, 3}
  my_set.add(4)
  # Result: {1, 2, 3, 4}
  my_set.remove(2)
  # Result: {1, 3, 4}
  my_set.discard(5)
  # No error, as 5 is not in the set
  popped_element = my_set.pop()
  # Result: popped_element contains the removed element
  ```



````python
my_set = {1, 2, 3}

my_set.add(4)
print('Add method',my_set)

my_set.remove(2)
print('Remove method',my_set)

my_set.discard(5)
print('Discard Method',my_set)

popped_element = my_set.pop()
print('Pop Method',popped_element )
````

## Stack Operations README

### Introduction to Stacks:
- A stack is a collection of items where interactions are limited to the top item.
- It follows the LIFO (Last-In, First-Out) principle.

### LIFO (Last-In, First-Out) Principle:
- The last item added is the first one to be removed.
- Visualize it like a stack of plates or books.

### Operations:
1. **push():**
   - Adds an item to the top of the stack.
   - Analogous to placing a book on top of a stack.

2. **pop():**
   - Removes the top item from the stack.
   - Similar to taking the top book off the stack.

3. **isEmpty():**
   - Checks if the stack has no items.
   - Like verifying if there are no books in the stack.

4. **isFull():**
   - Checks if the stack has reached its maximum capacity.
   - Similar to checking if you can add more books when the stack is full.

5. **peek():**
   - Views the top item without removing it.
   - Imagine checking the title of the top book without taking it off.

6. **count():**
   - Counts the number of items in the stack.
   - Similar to counting the books in the stack.

7. **change():**
   - Modifies the item at a specific position.
   - Like replacing a book in the middle of the stack.

8. **display():**
   - Shows all the items in the stack.
   - Similar to displaying all the books in the stack.

### Implementing Stacks Using Lists:
- In Python, a list can represent a stack.
- Use `append()` to push items onto the stack and `pop()` to remove items from the top.
- Check if the stack is empty with `if not stack:` and check if it's full by comparing the length with a predefined limit.

In summary, a stack organizes items with a "last in, first out" approach. The provided operations facilitate interactions and manipulations on the stack. Implementing a stack using lists in Python is a practical way to work with this concept in programming.



````python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            print("Stack is empty. Cannot pop.")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            print("Stack is empty. Cannot peek.")

    def size(self):
        return len(self.items)
````

````python
stack = Stack()

stack.push(1)
stack.push(2)
stack.push(3)

print("Stack:", stack.items)

print("Pop:", stack.pop())
print("Stack after pop:", stack.items)

print("Peek:", stack.peek())

print("Stack size:", stack.size())
````

## Queue Operations README

### Introduction to Queues:
- A queue is a linear data structure representing a line of elements.
- It follows the FIFO (First-In, First-Out) principle.
- Imagine it as people waiting in line, where the first person to arrive is the first to be served.

### FIFO (First-In, First-Out) Principle:
- The first item added to the queue is the first one to be removed.
- Analogous to the first person waiting in line being the first to get served.

### Operations:
1. **enqueue():**
   - Adds an item to the end of the queue.
   - Similar to joining the line of people waiting.

2. **dequeue():**
   - Removes the item from the front of the queue.
   - Like the person at the front of the line getting served and leaving.

3. **isEmpty():**
   - Checks if the queue has no items.
   - Similar to seeing if there is no one waiting in line.

4. **isFull():**
   - Checks if the queue has reached its maximum capacity.
   - Similar to checking if the line is so long that no more people can join.

5. **peek():**
   - Looks at the front item without removing it.
   - Imagine checking who is at the front of the line without asking them to leave.

6. **count():**
   - Counts how many items are in the queue.
   - Like counting the number of people waiting in line.

7. **change():**
   - Modifies the item at a specific position.
   - Similar to asking someone in the middle of the line to step aside briefly.

8. **display():**
   - Shows all the items in the queue.
   - Like displaying everyone in the line.

### Implementing Queues Using Lists or collections.deque:
- In Python, a list or `collections.deque` can represent a queue.
- To add an item, use the `append()` method, and to remove an item, use the `popleft()` method with `collections.deque`.
- Checking if the queue is empty is as simple as `if not queue:` and checking if it's full might involve comparing the length of the list or deque with a predefined limit.

In summary, a queue organizes items with a "first in, first out" approach. The provided operations facilitate interactions and manipulations on the queue. Implementing a queue using lists or `collections.deque` in Python is a practical way to work with this concept in programming.

Queue using a list

````python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise IndexError("dequeue from an empty queue")

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
````

````python
my_queue = Queue()
my_queue.enqueue(1)
my_queue.enqueue(2)
my_queue.enqueue(3)

print(my_queue.dequeue())
print(my_queue.size())
````

Queue using collection.deque

````python
from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.popleft()
        else:
            raise IndexError("dequeue from an empty queue")

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
````

````python
my_queue = Queue()
my_queue.enqueue(1)
my_queue.enqueue(2)
my_queue.enqueue(3)

print(my_queue.dequeue())
print(my_queue.size())
````

## Linked Lists
### Introduction to Linked Lists:
- A linked list is a dynamic data structure where each element (node) holds a value and a reference to the next node.
- Unlike arrays, linked lists don't require contiguous memory, allowing for dynamic size adjustments.

### Singly Linked Lists:
- In a singly linked list, each node points to the next node in the sequence.
- It's like a chain where each link only connects to the next link.

### Doubly Linked Lists:
- In a doubly linked list, each node has references to both the next and the previous nodes.
- Imagine a train where each carriage is connected to the one in front and behind.

### Circular Linked Lists:
- In a circular linked list, the last node points back to the first node, forming a loop.
- Picture a chain where the last link connects back to the first.

### Operations on Linked Lists:
1. **Insertion:**
   - *At the beginning:* Adding a new node at the start of the list.
   - *At the end:* Adding a new node at the end of the list.
   - *In the middle:* Adding a new node at a specific position within the list.

2. **Deletion:**
   - *At the beginning:* Removing the first node in the list.
   - *At the end:* Removing the last node in the list.
   - *In the middle:* Removing a node from a specific position within the list.

3. **Traversal:**
   - Moving through the linked list, visiting each node.
   - Starting from the first node and going until the last.

### Example Scenario - Insertion:
- *At the beginning:*
  - Adding a new station to the train line, becoming the first station.
- *At the end:*
  - Adding a new car to the end of the train.
- *In the middle:*
  - Inserting a new station between two existing stations.

### Example Scenario - Deletion:
- *At the beginning:*
  - Removing the first station from the train line.
- *At the end:*
  - Removing the last car from the train.
- *In the middle:*
  - Taking out a station from between two existing stations.

### Example Scenario - Traversal:
- Walking along the train track, visiting each station from the first to the last.

In summary, linked lists provide flexibility in organizing and managing data. Singly linked lists connect nodes in one direction, doubly linked lists connect nodes in both directions, and circular linked lists form a loop. Operations like insertion, deletion, and traversal allow you to manipulate the linked list based on your needs.

Single Linked List

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def delete_node(self, key):
        current = self.head
        if current and current.data == key:
            self.head = current.next
            return
        prev = None
        while current and current.data != key:
            prev = current
            current = current.next
        if current is None:
            return
        prev.next = current.next

    def display(self):
        current = self.head
        while current:
            print(current.data, end=' -> ')
            current = current.next
        print('None')

````

````python
linked_list = LinkedList()
linked_list.insert_at_end(1)
linked_list.insert_at_end(2)
linked_list.insert_at_end(3)
linked_list.display()
linked_list.delete_node(2)
linked_list.display()
````

Doubly Liked Lists

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def delete(self, key):
        current = self.head

        while current is not None and current.data != key:
            current = current.next

        if current is None:
            return

        if current.prev is not None:
            current.prev.next = current.next
        else:
            self.head = current.next

        if current.next is not None:
            current.next.prev = current.prev

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()
````

````python
dll = DoublyLinkedList()
dll.insert(1)
dll.insert(2)
dll.insert(3)
dll.display()
dll.delete(2)
dll.display()
````

Circular Linked List

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head

    def delete(self, key):
        current = self.head
        prev = None

        while current and current.data != key:
            prev = current
            current = current.next

        if current == self.head:
            prev = self.head
            while prev.next != self.head:
                prev = prev.next

        if current == self.head:
            if current.next == self.head:
                self.head = None
            else:
                prev.next = current.next
                self.head = current.next
        else:
            prev.next = current.next

    def display(self):
        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()
````

````python
cll = CircularLinkedList()
cll.insert(1)
cll.insert(2)
cll.insert(3)
cll.display()
cll.delete(2)
cll.display()
````

### Operations on Linked Lists:

Insertion:

````python
#Insertion at the Begining
def insert_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

#Insertion at the specific Position
def insert_at_position(self, data, position):
    new_node = Node(data)
    if position == 0:
        new_node.next = self.head
        self.head = new_node
    else:
        current = self.head
        for _ in range(position - 1):
            if current is None:
                raise Exception("Position out of bounds")
            current = current.next
        new_node.next = current.next
        current.next = new_node
````

Deletion:

````python
#Delete at the begining
def delete_at_beginning(self):
    if not self.head:
        raise Exception("List is empty")
    self.head = self.head.next

# Delete at a Specific Position
def delete_at_position(self, position):
    if not self.head:
        raise Exception("List is empty")
    current = self.head
    if position == 0:
        self.head = current.next
    else:
        for _ in range(position - 1):
            if current is None:
                raise Exception("Position out of bounds")
            current = current.next
        if current is None or current.next is None:
            raise Exception("Position out of bounds")
        current.next = current.next.next
````

## Trees and its Structures

### 1. Introduction to Trees:

Imagine a tree you see outside – it has a main trunk (the root) and branches coming out from it. These branches can further split into smaller branches or have leaves at the end. In computer science, a tree is a way to organize information in a similar hierarchical structure.





### 2. Binary Trees:

A binary tree is a special kind of structure where each "branch" or node can have at most two connections, usually called the left and right branches. It's like saying a parent can have at most two children.

### 3. Binary Search Trees:

Now, in a binary search tree (BST), each node not only has branches but also holds a value. The cool thing about a BST is that it's organized in a way that makes finding a particular value very efficient. If a value is smaller than the current node, you go to the left; if it's larger, you go to the right.

### 4. Tree Traversal Algorithms (In-order, Pre-order, Post-order):

Traversing a tree means visiting and checking each node. We have different ways to do this:

- **In-order Traversal:**
  Think of it like reading a book from left to right – you visit the left branch, then the current node, and finally the right branch.

- **Pre-order Traversal:**
  This is like starting at the top of a tree and moving down – you visit the current node first, then the left branch, and finally the right branch.

- **Post-order Traversal:**
  Imagine starting from the bottom of a tree and going up – you visit the left branch, then the right branch, and finally the current node.


image.png

````python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# In-order traversal
def inorder_traversal(node):
    if node:
        inorder_traversal(node.left)
        print(node.key, end=" ")
        inorder_traversal(node.right)

# Pre-order traversal
def preorder_traversal(node):
    if node:
        print(node.key, end=" ")
        preorder_traversal(node.left)
        preorder_traversal(node.right)

# Post-order traversal
def postorder_traversal(node):
    if node:
        postorder_traversal(node.left)
        postorder_traversal(node.right)
        print(node.key, end=" ")

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output traversals
print("In-order traversal:", end=" ")
inorder_traversal(root)
print()

print("Pre-order traversal:", end=" ")
preorder_traversal(root)
print()

print("Post-order traversal:", end=" ")
postorder_traversal(root)
print()
````

### 5. Tree Balancing (AVL Trees, Red-Black Trees):

Trees can get unbalanced if we keep adding nodes in a certain order, making operations slower. AVL trees and Red-Black trees are like tree caretakers that ensure the tree stays balanced.

- **AVL Trees:**

### 3. Graph Traversal Algorithms (Depth-First Search, Breadth-First Search):

Now, once you have a graph, you might want to explore it. Two common ways to do this are:

- **Depth-First Search (DFS):**
  Imagine you're exploring a maze. You go as far as you can down one path before backtracking. DFS does something similar – it explores as far as possible along one branch of the graph before backtracking.

- **Breadth-First Search (BFS):**
  Picture dropping a stone in a pond – the ripples spread out equally. BFS works in a similar way, exploring all the neighbors of a node before moving on to their neighbors.

So, if your graph represents, say, a network of roads, DFS might be like exploring one road until it ends, while BFS might be like checking all the roads that start from your current location before moving on.

In summary, graphs help model relationships, adjacency matrices and lists are ways to write down those relationships, and DFS and BFS are techniques to explore and understand the structure of the graph.

````python
from collections import deque

class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def bfs_traversal(root):
    if not root:
        return

    queue = deque([root])

    while queue:
        current = queue.popleft()
        print(current.key, end=" ")

        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output BFS traversal
print("BFS traversal:", end=" ")
bfs_traversal(root)
````

````python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# Recursive DFS
def dfs_recursive(node):
    if not node:
        return

    print(node.key, end=" ")
    dfs_recursive(node.left)
    dfs_recursive(node.right)

# Iterative DFS using a stack
def dfs_iterative(node):
    if not node:
        return

    stack = [node]

    while stack:
        current = stack.pop()
        print(current.key, end=" ")

        # Push right child first to visit it later (since stack is LIFO)
        if current.right:
            stack.append(current.right)

        # Push left child second to visit it earlier
        if current.left:
            stack.append(current.left)

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output DFS traversals
print("Recursive DFS:", end=" ")
dfs_recursive(root)
print()

print("Iterative DFS:", end=" ")
dfs_iterative(root)
print()
````

## Heaps
### 1. Introduction to Heaps:

A heap is a special kind of tree-like data structure where each node has a value, and the value of a node is smaller (for a min-heap) or larger (for a max-heap) than the values of its children. This structure allows us to efficiently find and manipulate the smallest or largest element in the heap.



### 2. Min-Heap and Max-Heap:

- **Min-Heap:**
  - In a min-heap, the smallest value is at the root.
  - Every parent node has a value smaller than or equal to its child nodes.
  - It's like a hierarchy where each parent is smaller than its children.

  image.png

- **Max-Heap:**
  - In a max-heap, the largest value is at the root.
  - Every parent node has a value greater than or equal to its child nodes.
  - It's like a hierarchy where each parent is larger than its children.

image.png


````python
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def push(self, value):
        heapq.heappush(self.heap, value)

    def pop(self):
        return heapq.heappop(self.heap)

# Example usage of MinHeap
min_heap = MinHeap()
min_heap.push(3)
min_heap.push(1)
min_heap.push(4)
min_heap.push(2)

print("Min-Heap elements:")
while min_heap.heap:
    print(min_heap.pop(), end=" ")
print()
````

````python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def push(self, value):
        heapq.heappush(self.heap, -value)

    def pop(self):
        return -heapq.heappop(self.heap)

# Example usage of MaxHeap
max_heap = MaxHeap()
max_heap.push(3)
max_heap.push(1)
max_heap.push(4)
max_heap.push(2)

print("Max-Heap elements:")
while max_heap.heap:
    print(max_heap.pop(), end=" ")
print()
````

### 3. Heap Operations:

- **Insert:**
  - To add a new element to the heap.
  - The new element is usually added at the bottom, maintaining the heap property.
  - After insertion, the heap might need to be rearranged to preserve the order.

- **Extract-Min (or Extract-Max):**
  - To remove and retrieve the smallest (for min-heap) or largest (for max-heap) element from the heap.
  - After extraction, the heap structure is adjusted to ensure it remains a valid heap.



````python
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        heapq.heappush(self.heap, value)

    def extract_min(self):
        if self.heap:
            return heapq.heappop(self.heap)
        else:
            return None

# Example usage of MinHeap with insert and extract_min
min_heap = MinHeap()

# Insert elements
min_heap.insert(3)
min_heap.insert(1)
min_heap.insert(4)
min_heap.insert(2)

print("Min-Heap elements:")
print(min_heap.heap)

# Extract-Min
min_value = min_heap.extract_min()
print(f"Extracted Min: {min_value}")
print("Min-Heap elements after extraction:")
print(min_heap.heap)
````

````python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        heapq.heappush(self.heap, -value)

    def extract_max(self):
        if self.heap:
            return -heapq.heappop(self.heap)
        else:
            return None

# Example usage of MaxHeap with insert and extract_max
max_heap = MaxHeap()

# Insert elements
max_heap.insert(3)
max_heap.insert(1)
max_heap.insert(4)
max_heap.insert(2)

print("Max-Heap elements:")
print(max_heap.heap)

# Extract-Max
max_value = max_heap.extract_max()
print(f"Extracted Max: {max_value}")
print("Max-Heap elements after extraction:")
print(max_heap.heap)
````

## Hashing
### 1. Introduction to Hashing:

Hashing is a technique used to efficiently retrieve, store, and manage data in a way that allows for quick access. It involves using a function called a hash function to map data to a fixed-size array or structure, typically called a hash table.



### 2. Hash Functions:

- **Definition:**
  - A hash function takes in data (like a key) and produces a fixed-size string of characters, which is usually a hash code.
  - The output of a hash function is a unique representation (hash) of the input data.
  - Ideally, a hash function should generate different hash codes for different inputs, but collisions (same hash for different inputs) can occur.



### 3. Hash Tables:

- **Definition:**
  - A hash table is a data structure that uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.
  - It allows for quick retrieval and storage of data by using the hash code as an index to locate the data in the table.
  - Hash tables are widely used in implementing associative arrays, sets, and caches.


image.png

### 4. Collision Resolution Strategies:

- **Chaining:**
  - When two different inputs produce the same hash code (a collision), chaining involves creating a linked list or another data structure at that location to store multiple values.
  - The colliding values are stored in the same bucket and can be accessed by traversing the linked list.

- **Open Addressing:**
  - In open addressing, when a collision occurs, the system looks for the next available slot in the hash table to place the colliding value.
  - This process is continued until an empty slot is found.
  - Common open addressing methods include linear probing, quadratic probing, and double hashing.



### Example:

Let's say we have a hash function that calculates the sum of the ASCII values of the characters in a name and takes the remainder when divided by the size of the hash table. If we have a hash table with 10 slots:

- Hash Function: `hash(name) = sum(ASCII values) % 10`

- Hash Table:
  ```
  Slot 0: [ ]
  Slot 1: [ ]
  ...
  Slot 9: [ ]
  ```

- If we want to store the name "Alice," the hash function gives us a value of, say, 3. So, we store "Alice" in Slot 3.

- If we later want to store "Bob," and the hash function also gives us 3, we handle the collision:
  - Chaining: Create a linked list at Slot 3 to store both "Alice" and "Bob."
  - Open Addressing: Look for the next available slot (maybe Slot 4, then Slot 5) until an empty slot is found for "Bob."

## Sorting Algorithms:

### 1. **Bubble Sort:**
   - Imagine you have a list of numbers, and you want to arrange them in order.
   - Bubble sort works by repeatedly going through the list, comparing two adjacent numbers, and swapping them if they are in the wrong order.
   - This process "bubbles up" the largest unsorted element to its correct position.


image.png

````python
def bubble_sort(arr):
    n = len(arr)
    steps = []

    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                steps.append(arr.copy())

    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    unsorted_list = [int(x) for x in input_list.split()]

    print("Unsorted list:", unsorted_list)

    # Perform Bubble Sort and get the steps
    sorting_steps = bubble_sort(unsorted_list.copy())

    # Display every step
    for step_num, step in enumerate(sorting_steps, start=1):
        print(f"Step {step_num}: {step}")

    # Display the final sorted list
    print("Sorted list:", unsorted_list)


if __name__ == "__main__":
    main()
````


### 2. **Selection Sort:**
   - Picture your list again. Selection sort finds the smallest (or largest) element and puts it in its proper place.
   - It does this by dividing the list into a sorted and an unsorted region. It selects the smallest (or largest) from the unsorted part and moves it to the sorted part.


image.png

````python
def selection_sort(arr):
    n = len(arr)
    steps = []

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap the found minimum element with the first element
        arr[i], arr[min_index] = arr[min_index], arr[i]
        steps.append(arr.copy())

    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    unsorted_list = [int(x) for x in input_list.split()]

    print("Unsorted list:", unsorted_list)

    # Perform Selection Sort and get the steps
    sorting_steps = selection_sort(unsorted_list.copy())

    # Display every step
    for step_num, step in enumerate(sorting_steps, start=1):
        print(f"Step {step_num}: {step}")

    # Display the final sorted list
    print("Sorted list:", unsorted_list)


if __name__ == "__main__":
    main()
````

### 3. **Insertion Sort:**
   - Think of sorting a deck of cards. You pick one card at a time and insert it into its correct position in the already sorted part of your hand.
   - Insertion sort does something similar. It builds the sorted part of the list by repeatedly taking an element from the unsorted part and inserting it into its correct place.

image.png

````python
def insertion_sort(arr):
    n = len(arr)
    steps = []

    for i in range(1, n):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
        steps.append(arr.copy())

    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    unsorted_list = [int(x) for x in input_list.split()]

    print("Unsorted list:", unsorted_list)

    # Perform Insertion Sort and get the steps
    sorting_steps = insertion_sort(unsorted_list.copy())

    # Display every step
    for step_num, step in enumerate(sorting_steps, start=1):
        print(f"Step {step_num}: {step}")

    # Display the final sorted list
    print("Sorted list:", unsorted_list)


if __name__ == "__main__":
    main()
````

### 4. **Merge Sort:**
   - Imagine splitting your list into halves repeatedly until you have many small lists. Then, merge these small lists back together in the correct order.
   - Merge sort is a "divide and conquer" algorithm. It divides the list into smaller parts, sorts them, and then merges them to get the final sorted list.


image.png

````python
def merge_sort(arr):
    steps = []

    def merge(left, right):
        result = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1

        result.extend(left[i:])
        result.extend(right[j:])
        return result

    def _merge_sort(arr):
        if len(arr) <= 1:
            return arr

        mid = len(arr) // 2
        left = _merge_sort(arr[:mid])
        right = _merge_sort(arr[mid:])
        sorted_arr = merge(left, right)
        steps.append(sorted_arr.copy())
        return sorted_arr

    _merge_sort(arr.copy())
    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    unsorted_list = [int(x) for x in input_list.split()]

    print("Unsorted list:", unsorted_list)

    # Perform Merge Sort and get the steps
    sorting_steps = merge_sort(unsorted_list.copy())

    # Display every step
    for step_num, step in enumerate(sorting_steps, start=1):
        print(f"Step {step_num}: {step}")

    # Display the final sorted list
    print("Sorted list:", sorting_steps[-1])


if __name__ == "__main__":
    main()
````

### 5. **Quicksort:**
   - Quicksort also uses a "divide and conquer" approach. It selects a "pivot" element and partitions the list into elements smaller and larger than the pivot.
   - The smaller and larger sections are then sorted independently. This process repeats until the entire list is sorted.


image.png

````python
def quick_sort(arr):
    steps = []

    def partition(low, high):
        pivot_index = (low + high) // 2
        pivot = arr[pivot_index]

        while low <= high:
            while arr[low] < pivot:
                low += 1
            while arr[high] > pivot:
                high -= 1

            if low <= high:
                arr[low], arr[high] = arr[high], arr[low]
                low += 1
                high -= 1

        return low

    def _quick_sort(low, high):
        if low < high:
            pivot = partition(low, high)
            _quick_sort(low, pivot - 1)
            _quick_sort(pivot, high)
            steps.append(arr.copy())

    _quick_sort(0, len(arr) - 1)
    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    unsorted_list = [int(x) for x in input_list.split()]

    print("Unsorted list:", unsorted_list)

    # Perform Quick Sort and get the steps
    sorting_steps = quick_sort(unsorted_list.copy())

    # Display every step
    for step_num, step in enumerate(sorting_steps, start=1):
        print(f"Step {step_num}: {step}")

    # Display the final sorted list
    print("Sorted list:", sorting_steps[-1])


if __name__ == "__main__":
    main()
````

## Time Complexity Analysis:

When we talk about the "time complexity," we're interested in how the algorithm's performance scales with the size of the input.

- **Bubble Sort, Selection Sort, Insertion Sort:**
   - These have a time complexity of O(n^2) in the worst case. It means that as the number of elements in your list increases, the time it takes to sort them quadratically increases.

- **Merge Sort and Quicksort:**
   - Merge sort and quicksort are more efficient, with a time complexity of O(n log n) in the average case.
   - This means their performance improves as the list gets longer, and they handle large datasets better than the quadratic time algorithms.

Understanding time complexity helps us choose the most efficient algorithm for a particular problem, especially when dealing with large sets of data. In general, we prefer algorithms with lower time complexity for better performance.

## Searching

## 1. **Linear Search:**
   - Imagine you have a list of items (numbers, names, etc.), and you're looking for a specific item.
   - Linear search is like checking each item in the list one by one until you find the one you're looking for.
   - It starts from the beginning and goes through each element until it finds a match or reaches the end of the list.


image.png

````python
def linear_search(arr, target):
    steps = []
    n = len(arr)

    for i in range(n):
        steps.append((i, arr[i]))

        if arr[i] == target:
            steps.append(f"Target {target} found at index {i}.")
            return steps

    steps.append(f"Target {target} not found in the list.")
    return steps


def main():
    # Get user input for the list
    input_list = input("Enter elements separated by space: ")
    search_list = [int(x) for x in input_list.split()]

    # Get user input for the target element
    target = int(input("Enter the target element to search for: "))

    print("List:", search_list)
    print("Target:", target)

    # Perform Linear Search and get the steps
    search_steps = linear_search(search_list, target)

    # Display every step
    for step_num, step in enumerate(search_steps, start=1):
        print(f"Step {step_num}: {step}")

if __name__ == "__main__":
    main()
````

## 2. **Binary Search:**
   - Picture a sorted list of items. Binary search is like playing a guessing game where you pick the middle item and determine whether the item you're looking for is before or after it.
   - If it's before, you repeat the process in the first half; if it's after, you repeat in the second half.
   - Binary search significantly reduces the number of elements you need to check compared to linear search, especially with large sorted lists.

image.png


````python
def binary_search(arr, target):
    steps = []
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        steps.append((low, high, mid, arr[mid]))

        if arr[mid] == target:
            steps.append(f"Target {target} found at index {mid}.")
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    steps.append(f"Target {target} not found in the list.")
    return steps


def main():
    # Get user input for the sorted list
    input_list = input("Enter sorted elements separated by space: ")
    sorted_list = [int(x) for x in input_list.split()]

    # Get user input for the target element
    target = int(input("Enter the target element to search for: "))

    print("Sorted List:", sorted_list)
    print("Target:", target)

    # Perform Binary Search and get the steps
    search_steps = binary_search(sorted_list, target)

    # Display every step
    for step_num, step in enumerate(search_steps, start=1):
        if isinstance(step, tuple):
            print(f"Step {step_num}: Low={step[0]}, High={step[1]}, Mid={step[2]}, Mid Element={step[3]}")
        else:
            print(f"Step {step_num}: {step}")

if __name__ == "__main__":
    main()
````

## 3. **Hash-Based Search:**
   - Think of a dictionary with words and their meanings. Hash-based search is like finding a word by directly going to the page where it should be based on the first letter.
   - In computer science, this involves using a hash function to map keys (the things you're searching for) to positions in a data structure (like an array).
   - This allows for very fast lookups, assuming a well-designed hash function and minimal collisions (when two keys map to the same position).



## Comparison:

- **Linear Search:**
   - Simple but can be slow, especially for large lists.
   - Works for any list, sorted or not.

- **Binary Search:**
   - Requires a sorted list but is much faster than linear search.
   - Efficient for large datasets but requires a sorted list.

- **Hash-Based Search:**
   - Fast for lookups but requires a good hash function.
   - Efficient for large datasets with well-designed hash functions.

## Summary:

- Use **Linear Search** when the list is small or not sorted.
- Use **Binary Search** when you have a sorted list.
- Use **Hash-Based Search** when you need fast lookups and have a way to efficiently map keys to positions.
