

## Scatter Plot:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for demonstration
np.random.seed(42)
x_scatter = np.random.rand(50)
y_scatter = 2 * x_scatter + 1 + 0.1 * np.random.randn(50)

# Creating a Scatter Plot
plt.scatter(x_scatter, y_scatter, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.legend()
plt.grid(True)
plt.show()
```



## Pie Chart:



```python
# Generating random data for demonstration
labels_pie = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
sizes_pie = [25, 30, 20, 25]

# Creating a Pie Chart
plt.pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```
