## Heaps
### 1. Introduction to Heaps:

A heap is a special kind of tree-like data structure where each node has a value, and the value of a node is smaller (for a min-heap) or larger (for a max-heap) than the values of its children. This structure allows us to efficiently find and manipulate the smallest or largest element in the heap.



### 2. Min-Heap and Max-Heap:

- **Min-Heap:**
  - In a min-heap, the smallest value is at the root.
  - Every parent node has a value smaller than or equal to its child nodes.
  - It's like a hierarchy where each parent is smaller than its children.

  image.png

- **Max-Heap:**
  - In a max-heap, the largest value is at the root.
  - Every parent node has a value greater than or equal to its child nodes.
  - It's like a hierarchy where each parent is larger than its children.

image.png


````python
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def push(self, value):
        heapq.heappush(self.heap, value)

    def pop(self):
        return heapq.heappop(self.heap)

# Example usage of MinHeap
min_heap = MinHeap()
min_heap.push(3)
min_heap.push(1)
min_heap.push(4)
min_heap.push(2)

print("Min-Heap elements:")
while min_heap.heap:
    print(min_heap.pop(), end=" ")
print()
````

````python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def push(self, value):
        heapq.heappush(self.heap, -value)

    def pop(self):
        return -heapq.heappop(self.heap)

# Example usage of MaxHeap
max_heap = MaxHeap()
max_heap.push(3)
max_heap.push(1)
max_heap.push(4)
max_heap.push(2)

print("Max-Heap elements:")
while max_heap.heap:
    print(max_heap.pop(), end=" ")
print()
````

### 3. Heap Operations:

- **Insert:**
  - To add a new element to the heap.
  - The new element is usually added at the bottom, maintaining the heap property.
  - After insertion, the heap might need to be rearranged to preserve the order.

- **Extract-Min (or Extract-Max):**
  - To remove and retrieve the smallest (for min-heap) or largest (for max-heap) element from the heap.
  - After extraction, the heap structure is adjusted to ensure it remains a valid heap.



````python
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        heapq.heappush(self.heap, value)

    def extract_min(self):
        if self.heap:
            return heapq.heappop(self.heap)
        else:
            return None

# Example usage of MinHeap with insert and extract_min
min_heap = MinHeap()

# Insert elements
min_heap.insert(3)
min_heap.insert(1)
min_heap.insert(4)
min_heap.insert(2)

print("Min-Heap elements:")
print(min_heap.heap)

# Extract-Min
min_value = min_heap.extract_min()
print(f"Extracted Min: {min_value}")
print("Min-Heap elements after extraction:")
print(min_heap.heap)
````

````python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        heapq.heappush(self.heap, -value)

    def extract_max(self):
        if self.heap:
            return -heapq.heappop(self.heap)
        else:
            return None

# Example usage of MaxHeap with insert and extract_max
max_heap = MaxHeap()

# Insert elements
max_heap.insert(3)
max_heap.insert(1)
max_heap.insert(4)
max_heap.insert(2)

print("Max-Heap elements:")
print(max_heap.heap)

# Extract-Max
max_value = max_heap.extract_max()
print(f"Extracted Max: {max_value}")
print("Max-Heap elements after extraction:")
print(max_heap.heap)
````

### 4. Priority Queues:

- A priority queue is a data structure that uses a heap to efficiently manage elements with priorities.
- Elements with higher priorities are processed before those with lower priorities.
- Priority queues are often used in scenarios where certain tasks or data need to be handled based on their urgency or importance.



#### Example:

Let's consider a min-heap:

```
          10
        /    \
      20      15
     /  \    /  
    25   30  40
```

- In this min-heap, 10 is the smallest element.
- If we insert 5, it goes to the next available spot, and the heap is adjusted to maintain the min-heap property.
- If we extract the minimum (5), the heap is rearranged, and the next minimum (10) becomes the root.

Understanding heaps is essential for efficiently managing and accessing priority-based data, and they find applications in various algorithms and data processing scenarios.

## Hashing
### 1. Introduction to Hashing:

Hashing is a technique used to efficiently retrieve, store, and manage data in a way that allows for quick access. It involves using a function called a hash function to map data to a fixed-size array or structure, typically called a hash table.



### 2. Hash Functions:

- **Definition:**
  - A hash function takes in data (like a key) and produces a fixed-size string of characters, which is usually a hash code.
  - The output of a hash function is a unique representation (hash) of the input data.
  - Ideally, a hash function should generate different hash codes for different inputs, but collisions (same hash for different inputs) can occur.



### 3. Hash Tables:

- **Definition:**
  - A hash table is a data structure that uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.
  - It allows for quick retrieval and storage of data by using the hash code as an index to locate the data in the table.
  - Hash tables are widely used in implementing associative arrays, sets, and caches.


image.png

### 4. Collision Resolution Strategies:

- **Chaining:**
  - When two different inputs produce the same hash code (a collision), chaining involves creating a linked list or another data structure at that location to store multiple values.
  - The colliding values are stored in the same bucket and can be accessed by traversing the linked list.

- **Open Addressing:**
  - In open addressing, when a collision occurs, the system looks for the next available slot in the hash table to place the colliding value.
  - This process is continued until an empty slot is found.
  - Common open addressing methods include linear probing, quadratic probing, and double hashing.



### Example:

Let's say we have a hash function that calculates the sum of the ASCII values of the characters in a name and takes the remainder when divided by the size of the hash table. If we have a hash table with 10 slots:

- Hash Function: `hash(name) = sum(ASCII values) % 10`

- Hash Table:
  ```
  Slot 0: [ ]
  Slot 1: [ ]
  ...
  Slot 9: [ ]
  ```

- If we want to store the name "Alice," the hash function gives us a value of, say, 3. So, we store "Alice" in Slot 3.

- If we later want to store "Bob," and the hash function also gives us 3, we handle the collision:
  - Chaining: Create a linked list at Slot 3 to store both "Alice" and "Bob."
  - Open Addressing: Look for the next available slot (maybe Slot 4, then Slot 5) until an empty slot is found for "Bob."
