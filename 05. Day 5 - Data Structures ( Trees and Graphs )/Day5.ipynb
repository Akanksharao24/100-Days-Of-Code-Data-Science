## Trees and its Structures

### 1. Introduction to Trees:

Imagine a tree you see outside â€“ it has a main trunk (the root) and branches coming out from it. These branches can further split into smaller branches or have leaves at the end. In computer science, a tree is a way to organize information in a similar hierarchical structure.





### 2. Binary Trees:

A binary tree is a special kind of structure where each "branch" or node can have at most two connections, usually called the left and right branches. It's like saying a parent can have at most two children.

image.png


### 3. Binary Search Trees:

Now, in a binary search tree (BST), each node not only has branches but also holds a value. The cool thing about a BST is that it's organized in a way that makes finding a particular value very efficient. If a value is smaller than the current node, you go to the left; if it's larger, you go to the right.


image.png

### 4. Tree Traversal Algorithms (In-order, Pre-order, Post-order):

Traversing a tree means visiting and checking each node. We have different ways to do this:

- **In-order Traversal:**
  Think of it like reading a book from left to right â€“ you visit the left branch, then the current node, and finally the right branch.

- **Pre-order Traversal:**
  This is like starting at the top of a tree and moving down â€“ you visit the current node first, then the left branch, and finally the right branch.

- **Post-order Traversal:**
  Imagine starting from the bottom of a tree and going up â€“ you visit the left branch, then the right branch, and finally the current node.


image.png

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# In-order traversal
def inorder_traversal(node):
    if node:
        inorder_traversal(node.left)
        print(node.key, end=" ")
        inorder_traversal(node.right)

# Pre-order traversal
def preorder_traversal(node):
    if node:
        print(node.key, end=" ")
        preorder_traversal(node.left)
        preorder_traversal(node.right)

# Post-order traversal
def postorder_traversal(node):
    if node:
        postorder_traversal(node.left)
        postorder_traversal(node.right)
        print(node.key, end=" ")

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output traversals
print("In-order traversal:", end=" ")
inorder_traversal(root)
print()

print("Pre-order traversal:", end=" ")
preorder_traversal(root)
print()

print("Post-order traversal:", end=" ")
postorder_traversal(root)
print()
```

### 5. Tree Balancing (AVL Trees, Red-Black Trees):

Trees can get unbalanced if we keep adding nodes in a certain order, making operations slower. AVL trees and Red-Black trees are like tree caretakers that ensure the tree stays balanced.

- **AVL Trees:**
  AVL trees make sure that the difference in height between the left and right branches of any node is at most one. This keeps the tree nice and balanced.


image.png
  

- **Red-Black Trees:**
  Red-Black trees use a clever way of coloring nodes to maintain balance. They have rules like "no two red nodes in a row" that, when followed, ensure the tree doesn't become lopsided

  image.png

So, in a nutshell, trees help us organize data, binary trees give a structure to that organization, binary search trees make searches efficient, traversal algorithms help us explore the tree, and balancing methods like AVL and Red-Black trees keep everything in order for optimal performance.


## Graphs
### 1. Introduction to Graphs:

Think of a graph like a map. It represents a set of points (vertices) connected by lines (edges). These connections could represent relationships, roads, or any interaction between the points. In computer science, graphs are a way to model complex relationships between different things.



image.png

### 2. Representing Graphs (Adjacency Matrix, Adjacency List):

Graphs can be represented in two main ways:

- **Adjacency Matrix:**
  It's like a table where each row and column represent a vertex, and the cell at the intersection tells you if there's a connection (edge) between those two vertices.


image.png


- **Adjacency List:**
  Think of it like a list where each vertex has a list of its neighboring vertices. If two vertices are connected, they're listed as neighbors.

  image.png

Imagine you have a group of friends, and an adjacency matrix would be like a table saying who is friends with whom, while an adjacency list would be like each person having a list of their friends.



### 3. Graph Traversal Algorithms (Depth-First Search, Breadth-First Search):

Now, once you have a graph, you might want to explore it. Two common ways to do this are:

- **Depth-First Search (DFS):**
  Imagine you're exploring a maze. You go as far as you can down one path before backtracking. DFS does something similar â€“ it explores as far as possible along one branch of the graph before backtracking.

- **Breadth-First Search (BFS):**
  Picture dropping a stone in a pond â€“ the ripples spread out equally. BFS works in a similar way, exploring all the neighbors of a node before moving on to their neighbors.

So, if your graph represents, say, a network of roads, DFS might be like exploring one road until it ends, while BFS might be like checking all the roads that start from your current location before moving on.

In summary, graphs help model relationships, adjacency matrices and lists are ways to write down those relationships, and DFS and BFS are techniques to explore and understand the structure of the graph.

```python
from collections import deque

class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def bfs_traversal(root):
    if not root:
        return

    queue = deque([root])

    while queue:
        current = queue.popleft()
        print(current.key, end=" ")

        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output BFS traversal
print("BFS traversal:", end=" ")
bfs_traversal(root)
```

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# Recursive DFS
def dfs_recursive(node):
    if not node:
        return

    print(node.key, end=" ")
    dfs_recursive(node.left)
    dfs_recursive(node.right)

# Iterative DFS using a stack
def dfs_iterative(node):
    if not node:
        return

    stack = [node]

    while stack:
        current = stack.pop()
        print(current.key, end=" ")

        # Push right child first to visit it later (since stack is LIFO)
        if current.right:
            stack.append(current.right)

        # Push left child second to visit it earlier
        if current.left:
            stack.append(current.left)

# Example tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Output DFS traversals
print("Recursive DFS:", end=" ")
dfs_recursive(root)
print()

print("Iterative DFS:", end=" ")
dfs_iterative(root)
print()
```
